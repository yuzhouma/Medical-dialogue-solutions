```python
import subprocess
import time
from datetime import timedelta
import os
import sys
'''
CUDA_VISIBLE_DEVICES=0 nohup bash -c \"swift sft --model Qwen/Qwen2.5-7B-Instruct --train_type lora --dataset '/mnt/workspace/final_data/train.json' --torch_dtype bfloat16 --quant_method hqq --quant_bits 4 --num_train_epochs 1 --per_device_train_batch_size 2 --per_device_eval_batch_size 1 --learning_rate 1e-4 --lora_rank 8 --lora_alpha 32 --target_modules all-linear --gradient_accumulation_steps 16 --eval_steps 50 --save_steps 50 --save_total_limit 2 --logging_steps 5 --max_length 1024 --output_dir output --system 'You are a helpful assistant.' --warmup_ratio 0.05 --dataloader_num_workers 4 --model_author swift --model_name swift-robot\" &
'''


def run_swift_sft():
    # 记录开始时间
    start_time = time.time()
    print(f"开始时间: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(start_time))}\n")
    
    # 构建要执行的命令列表
    command_list = [
        "swift", "sft",
        "--model", "Qwen/Qwen2.5-7B-Instruct",
        "--train_type", "lora",
        "--dataset", "/mnt/workspace/final_data/train.json",
        "--torch_dtype", "bfloat16",
        "--num_train_epochs", "2",
        "--per_device_train_batch_size", "1",
        "--learning_rate", "1e-4",
        "--lora_rank", "8",
        "--lora_alpha", "32",
        "--seed", "42",
        "--gradient_accumulation_steps", "16",
        "--eval_steps", "50",
        "--save_steps", "50",
        "--save_total_limit", "2",
        "--logging_steps", "5",
        "--max_length", "1024",
        "--output_dir", "output",
        "--warmup_ratio", "0.05",
        "--dataloader_num_workers", "4",
        "--model_author", "swift",
        "--model_name", "swift-robot"
    ]
    
    # 显示命令回显
    print("即将执行的命令:")
    print(f"CUDA_VISIBLE_DEVICES=0 { ' '.join(command_list) }\n")
    print("开始执行命令，实时日志如下：\n")
    
    try:
        # 设置环境变量
        env = dict(os.environ)
        env["CUDA_VISIBLE_DEVICES"] = "0"
        
        # 执行命令并实时输出日志
        process = subprocess.Popen(
            command_list,
            env=env,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            text=True,
            bufsize=1  # 行缓冲
        )
        
        # 实时读取并打印输出
        for line in process.stdout:
            print(line, end='')  # end='' 避免重复换行
            sys.stdout.flush()  # 强制刷新输出缓冲区
            
        # 等待进程完成并获取返回码
        return_code = process.wait()
        if return_code != 0:
            raise subprocess.CalledProcessError(return_code, command_list)
        
    except subprocess.CalledProcessError as e:
        print(f"\n\n命令执行出错，返回码: {e.returncode}")
        return
    except Exception as e:
        print(f"\n\n发生异常: {e}")
        return
    
    # 记录结束时间
    end_time = time.time()
    print(f"\n\n结束时间: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(end_time))}")
    
    # 计算并输出时间差
    elapsed_time = end_time - start_time
    print(f"总执行时间: {timedelta(seconds=int(elapsed_time))}")

if __name__ == "__main__":
    run_swift_sft()
    
```
