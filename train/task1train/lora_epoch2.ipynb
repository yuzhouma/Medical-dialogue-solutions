```python
import subprocess
import time
from datetime import timedelta
import os
import sys
'''
CUDA_VISIBLE_DEVICES=0 nohup bash -c \"swift sft --model Qwen/Qwen2.5-7B-Instruct --train_type lora --dataset '/mnt/workspace/final_data/train.json' --torch_dtype bfloat16 --quant_method hqq --quant_bits 4 --num_train_epochs 1 --per_device_train_batch_size 2 --per_device_eval_batch_size 1 --learning_rate 1e-4 --lora_rank 8 --lora_alpha 32 --target_modules all-linear --gradient_accumulation_steps 16 --eval_steps 50 --save_steps 50 --save_total_limit 2 --logging_steps 5 --max_length 1024 --output_dir output --system 'You are a helpful assistant.' --warmup_ratio 0.05 --dataloader_num_workers 4 --model_author swift --model_name swift-robot\" &
'''


def run_swift_sft():
    # è®°å½•å¼€å§‹æ—¶é—´
    start_time = time.time()
    print(f"å¼€å§‹æ—¶é—´: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(start_time))}\n")
    
    # æ„å»ºè¦æ‰§è¡Œçš„å‘½ä»¤åˆ—è¡¨
    command_list = [
        "swift", "sft",
        "--model", "Qwen/Qwen2.5-7B-Instruct",
        "--train_type", "lora",
        "--dataset", "/mnt/workspace/final_data/train.json",
        "--torch_dtype", "bfloat16",
        "--num_train_epochs", "2",
        "--per_device_train_batch_size", "1",
        "--learning_rate", "1e-4",
        "--lora_rank", "8",
        "--lora_alpha", "32",
        "--seed", "42",
        "--gradient_accumulation_steps", "16",
        "--eval_steps", "50",
        "--save_steps", "50",
        "--save_total_limit", "2",
        "--logging_steps", "5",
        "--max_length", "1024",
        "--output_dir", "output",
        "--warmup_ratio", "0.05",
        "--dataloader_num_workers", "4",
        "--model_author", "swift",
        "--model_name", "swift-robot"
    ]
    
    # æ˜¾ç¤ºå‘½ä»¤å›æ˜¾
    print("å³å°†æ‰§è¡Œçš„å‘½ä»¤:")
    print(f"CUDA_VISIBLE_DEVICES=0 { ' '.join(command_list) }\n")
    print("å¼€å§‹æ‰§è¡Œå‘½ä»¤ï¼Œå®æ—¶æ—¥å¿—å¦‚ä¸‹ï¼š\n")
    
    try:
        # è®¾ç½®ç¯å¢ƒå˜é‡
        env = dict(os.environ)
        env["CUDA_VISIBLE_DEVICES"] = "0"
        
        # æ‰§è¡Œå‘½ä»¤å¹¶å®æ—¶è¾“å‡ºæ—¥å¿—
        process = subprocess.Popen(
            command_list,
            env=env,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            text=True,
            bufsize=1  # è¡Œç¼“å†²
        )
        
        # å®æ—¶è¯»å–å¹¶æ‰“å°è¾“å‡º
        for line in process.stdout:
            print(line, end='')  # end='' é¿å…é‡å¤æ¢è¡Œ
            sys.stdout.flush()  # å¼ºåˆ¶åˆ·æ–°è¾“å‡ºç¼“å†²åŒº
            
        # ç­‰å¾…è¿›ç¨‹å®Œæˆå¹¶è·å–è¿”å›ç 
        return_code = process.wait()
        if return_code != 0:
            raise subprocess.CalledProcessError(return_code, command_list)
        
    except subprocess.CalledProcessError as e:
        print(f"\n\nå‘½ä»¤æ‰§è¡Œå‡ºé”™ï¼Œè¿”å›ç : {e.returncode}")
        return
    except Exception as e:
        print(f"\n\nå‘ç”Ÿå¼‚å¸¸: {e}")
        return
    
    # è®°å½•ç»“æŸæ—¶é—´
    end_time = time.time()
    print(f"\n\nç»“æŸæ—¶é—´: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(end_time))}")
    
    # è®¡ç®—å¹¶è¾“å‡ºæ—¶é—´å·®
    elapsed_time = end_time - start_time
    print(f"æ€»æ‰§è¡Œæ—¶é—´: {timedelta(seconds=int(elapsed_time))}")

if __name__ == "__main__":
    run_swift_sft()
    
```
